# Feature Importance

```python
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score

import numpy as np

predictors=["age","gender","height","weight","ap_hi","ap_lo","cholesterol","gluc","smoke","alco","active"]
rf = RandomForestClassifier(random_state=1, n_estimators=50, min_samples_split=2, min_samples_leaf=5)
rf.fit(df[predictors],df["cardio"])

predictions = cross_val_predict(rf, df[predictors],df["cardio"])
predictions = pd.Series(predictions)
scores = cross_val_score(rf, df[predictors], df["cardio"],scoring='accuracy', cv=15)
print(scores)
importances=rf.feature_importances_
std = np.std([rf.feature_importances_ for tree in rf.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]
sorted_important_features=[]
for i in indices:
    sorted_important_features.append(predictors[i])
#predictors=titanic.columns
plt.figure()
plt.title("Feature Importances By Random Forest Model")
plt.bar(range(np.size(predictors)), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(np.size(predictors)), sorted_important_features, rotation='vertical')

plt.xlim([-1, np.size(predictors)]);
```

# Changing the outliers

```python
# Replacing all the 'ap_hi' values that are lower than 0 to mean value.
# Replacing all the 'ap_hi' values that are above 220 to standard deviation.

#Systolic blood pressure
for i in range (len(df['ap_hi'])):
    if df['ap_hi'][i]<0:
        df['ap_hi'][i]=df['ap_hi'].mean()
    elif df['ap_hi'][i]> 220:
        df['ap_hi'][i]=df['ap_hi'].std()
    else:
        pass
    
```

```python
# Replacing all the 'ap_lo' values that are lower than 0 to mean value.
# Replacing all the 'ap_lo' values that are above 140 to standard deviation.

# Diastolic Blood Pressure
for i in range (len(df['ap_lo'])):
    if df['ap_lo'][i]<0:
        df['ap_lo'][i]=df['ap_lo'].mean()
    elif df['ap_lo'][i]> 140:
        df['ap_lo'][i]=df['ap_lo'].std()
    else:
        pass
    
```

```python
# Age(in days) To Age(in years)
import numpy as np
for i in range(len(df['age'])):
    df['age'][i]=df['age'][i]/365

df['age']=np.floor(df['age'])
```

# Performing the T-Test to compare the blood pressure

```python
from scipy.stats import ttest_ind
with_cardio = df[df['cardio'] == 1]
without_cardio = df[df['cardio'] == 0]

t_stat, p_value = ttest_ind(with_cardio['ap_hi'], without_cardio['ap_hi'])
print(f'T-statistic: {t_stat}, p-value: {p_value}')
```

### Results of T Test

T-Stats showed us that there is substantial difference in the two groups of cardio vascular disorder thus it makes it viable to use for the machine learning model
